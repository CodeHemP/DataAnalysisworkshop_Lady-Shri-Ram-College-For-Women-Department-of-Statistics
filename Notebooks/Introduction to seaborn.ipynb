**Seaborn Library:** It is a library that provides a high-level interface to the matplotlib library for the creation of plots. The library is designed to make visualization of data easy and intuitive. To use seaborn, it is required to import matlotlib along with seaborn

```python
#import libraries
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
#plotting distplot with seaborn using random data to visualize the distribution of the data.
sns.set(style="white")
rs = np.random.RandomState(10)
d= rs.normal(size=100)
sns.distplot(d, kde=True, color="blue",bins=20)
```

```python
#x axis values
x = ['sun', 'mon', 'tue', 'wed', 'thu', 'fri', 'sat']
# y axis values
y= [10, 6, 2, 1, 4, 5, 8]
#plotting strip plot using seaborn
ax = sns.stripplot(x,y);
#giving labels to the x and y axis
ax.set(xlabel='Day of the week', ylabel='Number of people')
#to show the plot
plt.show()
```

```python
x = ['Male', 'Female']
y = [20, 15]
ax = sns.barplot(x,y);
ax.set(xlabel='Sex', ylabel='Count in office')
plt.show()
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=df)
plt.show()
```

```python
#plotting a lineplot with iris data for its two columns sepal length and sepal width.
sns.set(style="ticks")
#df = sns.load_dataset("iris")
sns.lmplot(x="sepal_length", y="sepal_width",data=df, size=6,aspect=1)
```

```python
#plotting a lineplot with iris data for its two columns sepal length and sepal width and be defining the hue as the species.
sns.set(style="ticks")
#df = sns.load_dataset("iris")
sns.lmplot(x="sepal_length", y="sepal_width", hue="species" ,data=df, size=6,aspect=1)
```

```python
#plotting strip plot with iris data for its two columns sepal length and species using seaborn.
#use to set the style of backgroud of the plot
sns.set(style='whitegrid')
#df = sns.load_dataset("iris")
ax = sns.stripplot(x="species", y="sepal_length", data=df, jitter=True)
plt.title('Strip Plot')
plt.show()
```
